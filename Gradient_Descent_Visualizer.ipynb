```python
import numpy as np
import matplotlib.pyplot as plt

# Our simple function: f(x) = x^2 + 3x + 2
def f(x):
    return x**2 + 3*x + 2

# Derivative: f'(x) = 2x + 3
def df(x):
    return 2*x + 3

# Gradient Descent Settings
x = 5          # starting point
lr = 0.1       # learning rate (step size)
history = []   # to store the path

for i in range(25):
    history.append(x)
    x = x - lr * df(x)   # move downhill

# Plot curve
xs = np.linspace(-6, 2, 200)
ys = f(xs)

plt.plot(xs, ys, label="f(x) curve")
plt.scatter(history, f(np.array(history)), color="red", label="descent path")
plt.title("Gradient Descent Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

print("Minimum found near x =", x)
```
